```python
import os
import tweepy
import pandas as pd
```

## Tweepy

Para acessar a API do twitter, deve-se acessar a √°rea de desenvolvedor, criar uma aplica√ß√£o e solicitar as credenciais de acesso. As c√©lulas abaixo demonstram como fazer a conex√£o na API do twitter usando a biblioteca python Tweepy e retornar a lista de tweets na timeline do usu√°rio.

[Documenta√ß√£o do tweepy](https://tweepy.readthedocs.io/)

```python
consumer_key = os.environ['CONSUMER_API_KEY']
consumer_secret = os.environ['CONSUMER_API_SECRET']
access_token = os.environ['ACCESS_TOKEN']
access_token_secret = os.environ['ACCESS_TOKEN_SECRET']
```

```python
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)

api = tweepy.API(auth)
```

Recuperar tweets da timeline

```python
public_tweets = api.home_timeline()
```

Criar um dicion√°rio vazio para incluir os tweets.
Passo necess√°rio para criar um dataframe.

```python
tweets = {'id':[],'author':[],'tweet':[],'created_at':[],'language':[],'n_retweets':[],'n_likes':[]}
```

```python
for tweet in public_tweets:
    print("id: "+tweet.id_str)
    print("Autor: "+tweet.author.name)
    print("Tweet: "+tweet.text)
    print("Data de cria√ß√£o: "+tweet.created_at.strftime("%d/%m/%Y, %H:%M:%S"))
    print("Idioma: "+tweet.lang)
    print("Retweets: "+str(tweet.retweet_count))
    print("Curtidas: "+str(tweet.favorite_count))
    tweets['id'].append(tweet.id)
    tweets['author'].append(tweet.author.name)
    tweets['tweet'].append(tweet.text)
    tweets['created_at'].append(tweet.created_at)
    tweets['language'].append(tweet.lang)
    tweets['n_retweets'].append(tweet.retweet_count)
    tweets['n_likes'].append(tweet.favorite_count)
    print("==========================")
```

Cria√ß√£o do dataframe

```python
df = pd.DataFrame.from_dict(tweets)
```
